```python
x = np.arange(0.0, 5.0, 0.2)
y1 = x
y2 = 0.25 * x ** 2
y3 = np.cos(x)

plt.plot(x, y1, '-o')
plt.plot(x, y2, '--s')
plt.plot(x, y3, ':x')
plt.legend(['y = x', 'y = 0.25 * x ** 2', 'y = cos(x)'])
plt.show()
```

```python
x = np.arange(0.0, 5.0, 0.2)
y1 = x
y2 = 0.25 * x ** 2
y3 = np.cos(x)

plt.plot(x, y1, '-o', label = 'r$y = x$')
plt.plot(x, y2, '--s', label = 'r$y = \frac{1]{4}x^2$')
plt.plot(x, y3, ':x', label = 'r$y = cos(x)$')
plt.title('Plot Tutorial')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X ** 2 + Y ** 2)
Z = np.sin(R)

import matplotlib.pyplot as plt

fig = plt.figure(figsize = (5, 5))
plt.contour(X, Y, Z, cmap = 'coolwarm')
plt.grid()
plt.axis('equal')
plt.show()
```
