```python
import random
```

```python
hangul = list("가나다라마바사아자차카타파하")
```

```python
with open("info.txt", "w") as file:
    for i in range(1000):
        name = random.choice(hangul) + random.choice(hangul)
        weight = random.randrange(40, 100)
        height = random.randrange(140, 200)
        file.write("{}, {}, {}\n". format(name, weight, height))
```

```python
f = open("info.txt", "r")
```
