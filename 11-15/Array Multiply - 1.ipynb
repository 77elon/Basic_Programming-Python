```python
x = 2 * np.random.rand(100, 1)
y = 4 + 3 * x + np.random.randn(1, 100)
```

```python
x_b = np.c_[np.ones((100, 1)), x]
```

```python
theta_best = np.linalg.inv(x_b.T.dot(x_b)).dot(x_b.T).dot(y)
```

```python
theta_best
```

```python
x_new = np.array([[0], [2]])
x_new_b = np.c_[np.ones((2, 1)), x_new]
y_predict = x_new_b.dot(theta_best)
y_predict
```

```python
plt.plot(x_new, y_predict, 'r-')
plt.plot(x, y, 'b.')
plt.axis([0, 2, 0, 15])
plt.show()
```

```python
w = tf.Variable([0.3], dtype=tf.float32)
b = tf.Variable([-0.3], dtype = tf.float32)
x = tf.placeholder(tf.float32)
linear_model = x * w + b
```

```python
y = tf.placeholder(tf.float32)
loss = tf.reduce_sum(tf.square(linear_model - y))
optimizer = tf.train.GradientDescentOptimizer(0.01)
train = optimizer.minimize(loss)
```

```python
x_train = [[1], [2], [3], [4]]
y_train = [[0.1], [-0.8], [-1.95], [-3]]
```

```python
MaxIter = 1000
init = tf.global_variables_initializer()
```

```python
with tf.compat.v1.Session() as sess:
    sess.run(init)
    for i in range(MaxIter):
        curr_w, curr_b, curr_loss = sess.run([w, b, loss], feed_dict={x:x_train, y:y_train})
        print(i, curr_w, curr_b, curr_loss)
        sess.run(train, feed_dict={x:x_train, y:y_train})
    curr_w, curr_b, curr_loss = sess.run([w, b, loss], feed_dict={x: x_train, y: y_train})
```

```python
x_train = np.array([1, 2, 3, 4])
y_train = np.array([0, -1, -2, -3])
```

```python
def l2_loss(w, b, x, y):
    return np.sum(np.square(w * x + b - y))
```

```python
def visualize_l2(w, b, x_train, y_train, loss):
    loss_name = l2_loss.__name__
    plt.plot(x_train, y_train, 'ok')
    plt.plot(x_train, w * x_train + b, '--kx')
    [plt.text(x, y, '($x_{0}$, $y_{0}$) = ({1}, {2})'.format(k, x, y), horizontalalignment = 'right')
    for k, (x, y) in enumerate(zip(x_train, y_train), 1)
    ]
    CurrentAxis = plt.gca()
    for xx, yy in zip(x_train, y_train):
        CurrentAxis.add_patch(Rectangle((xx, yy), (w * xx + b - yy), (w * xx + b - yy), alpha = 0.3, facecolor = 'gray', edgecolor='k'))
    plt.grid()
    plt.axis('equal')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('w = {}, b = {} ({} = {:1.2f})'.format(w, b, loss_name, l2_loss(w, b, x_train, y_train)))
    plt.show()
```

```python
plt.figure(figsize=(15, 9))
plt.subplot(2, 2, 1)
visualize_l2(0.2, 1, x_train, y_train, loss)
```

```python
plt.subplot(2, 2, 1)
visualize_l2(-1.0, 1.0, x_train, y_train, l2_loss)
```
