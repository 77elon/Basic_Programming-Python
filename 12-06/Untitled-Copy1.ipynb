```python
import numpy as np
import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import optimizers
```

```python
x_train = np.genfromtxt('train.in', dtype=np.float32)
```

```python
y_train = np.genfromtxt('train.out', dtype=np.float32)
```

```python
x_test = np.genfromtxt('result.in', dtype=np.float32)
```

```python
y_test = np.genfromtxt('result.out', dtype=np.float32)
```

```python
x_train.reshape(-1, 1)
```

```python
y_train = y_train.reshape(-1, 1)
y_test = y_test.reshape(-1, 1)
```

```python
y_train
```

```python
tf.model = tf.keras.Sequential()
tf.model.add(tf.keras.layers.Dense(units = 1, input_dim=4))
tf.model.add(tf.keras.layers.Activation('sigmoid'))  # this line can be omitted, as linear activation is default
tf.model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = tf.model.fit(x_train, y_train, epochs=100)
tf.model.summary()
```

```python
y_predict = tf.model.predict(x_test)
y_predict = tf.cast(y_predict >= 0.5, dtype=tf.float32)
```

```python
print((y_predict.numpy() == y_test).all())
```

```python
print(y_predict.numpy())
```

```python
meval_test = tf.model.evaluate(x_test, y_test)
print(meval_test) 
```

```python
meval = tf.model.evaluate(x_train, y_train)
print(meval) 
```
