```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
```

```python
x_train = np.genfromtxt('train.in')
```

```python
y_train = np.genfromtxt('train.out')
```

```python
x_test = np.genfromtxt('result.in')
```

```python
y_test = np.genfromtxt('result.out')
```

```python
y_train = y_train.reshape(-1, 1)
```

```python
y_train
```

```python
model = Sequential([
    Dense(units = 1, input_dim=4),  # input_shape=(2,)
    Activation('sigmoid')
])

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(x_train, y_train, epochs=10, batch_size=1, verbose=1)
```

```python
meval = model.evaluate(x_train, y_train)
print(meval)            # [0.209698(loss),  1.0(정확도)]

pred = model.predict(x_train)
print('예측 결과 : ', pred)     # [[0.16490099] [0.9996613 ]]
print('예측 결과 : ', np.squeeze(np.where(pred > 0.5, 1, 0)))  # [0 1]

for i in pred:
    if i > 0.5:
        print(1) 
    else:
        print(0)
```
